```python
from bs4 import BeautifulSoup
import urllib.request
response= urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')
html=response.read()
```

```python
soup=BeautifulSoup(html,"html5lib")
text=soup.get_text(strip=True)
tokens = [t for t in text.split()]
```

```python
freq = nltk.FreqDist(tokens)
for key,val in freq.items():
    if val > 5:
        print (str(key)+':'+str(val))
```

```python
freq.plot(20,cumulative=False)

```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

```

```python
stopwords.words('english')
clean_tokens=tokens[:]
sr=stopwords.words('english')

```

```python
from collections import Counter
dictionary=Counter(freq)
import matplotlib.pyplot as plt
from wordcloud import WordCloud

cloud = WordCloud(max_font_size=80,colormap="hsv").generate_from_frequencies(dictionary)
plt.figure(figsize=(16,12))
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
word_s= nltk.FreqDist({key:val for key,val in freq.items() if val > 100})
plt.pie(word_s.values(),explode=None,radius=2,labels=word_s.keys(),shadow=False)
plt.show()
```
